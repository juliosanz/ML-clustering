```python
from nltk.corpus import words
from nltk.corpus import stopwords
import os

e_stopwords = stopwords.words('english')
e_words = words.words()
print(len(e_words))

tokens_s = list()
for d in tokens:
    token_s = [w for w in d if not w in set(e_stopwords) and w.isalnum()]
    tokens_s.append(token_s)
```

```python
long_words = list()
for token in tokens_s:
    long_word = [w for w in token if len(w) > 15]
    long_words.append(long_word)
```

```python
from nltk.probability import FreqDist

all_long_tokens_list = [w for d in long_words for w in d]
all_long_tokens_set = set(all_long_tokens_list)
all_long_tokens_freq = FreqDist(all_long_tokens_list)
```

```python
NUMBER_OF_WORDS = 1340
tokens_to_search = [freq[0] for freq in all_long_tokens_freq.most_common(NUMBER_OF_WORDS)]
```

```python
from nltk.stem import WordNetLemmatizer

wl = WordNetLemmatizer()
tokens_l = list()
tokens_l = [wl.lemmatize(w, pos='v') for w in tokens_to_search]
```

```python
tokens_l2 = [wl.lemmatize(w, pos='a') for w in tokens_l]
```

```python
frequencies = list()
for row in long_words:
    frequencies.append(FreqDist(row))
```

```python
frequencies_calc = list()
for index in range(len(frequencies)):
    f = frequencies[index]
    if dict(f):
        frequencies_calc.append((f, index))
```
